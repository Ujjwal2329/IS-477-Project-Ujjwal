```python
import pandas as pd

sales_path = "../data/raw/coffee_sales.csv"
shop_path = "../data/raw/coffee_shop.csv"

sales = pd.read_csv(sales_path)
shop = pd.read_csv(shop_path)

print("Sales shape:", sales.shape)
print("Shop shape:", shop.shape)

sales.head()
```

```python
shop.head()
```

```python
print("=== Sales info ===")
print(sales.info())

print("\n=== Missing values in sales ===")
print(sales.isna().sum())

print("\n=== Shop info ===")
print(shop.info())

print("\n=== Missing values in shop ===")
print(shop.isna().sum())
```

```python
# Example summary: sales by store location
if "store_location" in sales.columns:
    print("\nSales count by store_location:")
    print(sales["store_location"].value_counts().head())

# Example summary: sales by product_category
if "product_category" in sales.columns:
    print("\nSales count by product_category:")
    print(sales["product_category"].value_counts().head())

# Example summary: money distribution in shop dataset
if "money" in shop.columns:
    print("\nMoney summary in coffee_shop:")
    print(shop["money"].describe())
```

```python
# For now, just save as cleaned copies (you can add actual cleaning later)
sales.to_csv("../data/processed/coffee_sales_clean.csv", index=False)
shop.to_csv("../data/processed/coffee_shop_clean.csv", index=False)
```
